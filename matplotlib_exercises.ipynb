```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Use matplotlib to plot the following equation:

y=x2−x+2

You'll need to write the code that generates the x and y points.

```python
x = np.array(range(0,10))
y = [(num**2)-(num+2) for num in x]

plt.plot(y)

plt.title ('$y = x^2 - x + 2$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.annotate('Origin', xy = (0,0), xytext = (0, 30), arrowprops = {})
```

Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

y=√x  
y=x^3  
y=2^x  
y=1/(x+1)  
You can use functions from the math module to help implement some of the equations above.

```python
x = np.array(range(0,10))
y = [num **(0.5) for num in x]

plt.plot(x,y)

plt.title ('$y=√x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
y = [num**3 for num in x]
plt.plot(x,y)

plt.title ('$y=x^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
y = [2**num for num in x]
plt.plot(y)
plt.title ('$y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
y=[1/(num+1) for num in x]
plt.plot(y)
plt.title ('$y= \dfrac{1}{x+1}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.subplot(221)
y = x **(0.5)
plt.plot(y)
plt.title ('$y=√x$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(222)
y=1/(x+1)
plt.plot(y)
plt.title ('$y= \dfrac{1}{x+1}$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(223)
y = 2**x
plt.plot(y)
plt.title ('$y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(224)
y = x**3
plt.plot(y)
plt.title ('$y=x^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.tight_layout()
```

Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
y = x**3
z = 2**x
j = 1/(x+1)
k = x **(0.5)


plt.plot(x,y, c='#30a2da')
plt.plot(x,z, c='firebrick')
plt.plot(x,k)
plt.plot(x,j)
plt.legend(['y','z','k','j'])
plt.suptitle('All Four One')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.ylim(0,10)
plt.xlim(0,10)
```
