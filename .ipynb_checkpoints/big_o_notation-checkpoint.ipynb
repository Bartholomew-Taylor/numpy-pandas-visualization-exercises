```python
x = range(1,10)

y = [0*n + 1 for n in x]
z = [n for n in x]
a = [n ** 2 for n in x]
b = [2**n for n in x]
c = [n**n for n in x]
d = [math.log(n) for n in x]
e = [math.factorial(n) for n in x]

plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,a)
plt.plot(x,b)
plt.plot(x,c)
plt.plot(x,d)
plt.plot(x,e)

plt.ylim(0,50)

plt.title('Range of Big O Nottation')
plt.legend(['y','z','a','b','c','d','e'])
```
