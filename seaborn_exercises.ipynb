Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
import os
import env 

from pydataset import data
import pandas as pd
import numpy as np
from env import host, username, password 
```

```python
  
    



def get_db_url(db, env_file = os.path.exists('env.py')):
    '''
    this will return a formatted string containing username, password, host, and database
    for connecting to the mysql server and the database indicated.
    env_file checks to see if the env.py exists in the cwd
    
    '''
    if env_file:
        username, password, host = (env.username, env.password, env.host)
        return f'mysql+pymysql://{username}:{password}@{host}/{db}'
    else:
        return 'you need to a username'
    
url = get_db_url('chipotle')
query = 'SELECT quantity, item_name, item_price FROM orders' 
top_chip = pd.read_sql(query, url)
top_chip.head()
```

```python
prices = top_chip['item_price'].str.strip("$").astype('float')
prices
```

```python
top_chip['price_actual'] = prices
top_chip
```

```python
bar_output = top_chip.groupby('item_name')[['price_actual']].agg('sum').sort_values(by = 'price_actual').tail(4)
bar_output
```

```python
bar_output = bar_output.reset_index()
```

```python
sns.barplot(y = 'price_actual', x = 'item_name', data = bar_output)
```
